```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic
```

```python
lst_features = [['Yes', 'Plasebo'], ['Yes', 'Aspirin'], ['No', 'Plasebo'], ['No', 'Aspirin']]
lst_num = [18, 7, 6, 13]
observed=[]
for features, num in zip(lst_features, lst_num):
  observed.extend([features for i in range(num)])

df = pd.DataFrame(observed, columns=['Thrombosis', 'Group'])
```

```python
print(pd.pivot_table(df, index=["Group"], columns=["Thrombosis"], aggfunc=lambda x: len(x)))

mosaic(df, ["Thrombosis","Group"], gap=0.01)
plt.show()
```
