```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_s = 5  # degreeses of freedom
x = np.linspace(0, 10, 100)
```

```python
for i in range(df_s):
  df = i + 1
  plt.plot(x, stats.chi2.pdf(x, df), label=f'k={df}')
plt.legend()
plt.title('Chi-square distribution with different degreeses of freedom')
plt.grid()
plt.show()
```
