```python
import seaborn as sns
import random
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
import warnings
```

```python
n = int(input('Number of experiments: '))
nt = int(input('Number of throws: '))
expect = [nt / 2] * 2
chi_res = []
p_res = []

for exp in range(n):

    coin_throw = np.random.randint(0, 2, nt)
    result = (np.count_nonzero(coin_throw == 0), np.count_nonzero(coin_throw == 1))
    chi, p = stats.chisquare(result, f_exp=expect)
    chi_res.append(chi)
    p_res.append(p)
```

```python
warnings.simplefilter('ignore')
fig, ax = plt.subplots(1, 2, figsize=(12,5))
sns.distplot(chi_res, ax=ax[0], color='red')
ax[0].set_xlim(0, max(chi_res))
sns.distplot(p_res, ax=ax[1], color='blue', bins=6)
ax[1].set_xlim(0, 1)
```

```python
np.mean(chi_res)
```
