```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
data = pd.read_csv('/content/Week4 - 11-Aug-to-17-Aug-2023.csv')
```

```python
data.head()
```

```python
data.columns
```

```python
data.describe()
```

```python
# for example
# Billie Eilish has two mentions as an artist
data[data['Artist'] == 'Billie Eilish'].head()
```

```python
artists = [i for i in data['Artist']]
x = [i for i in data['Artist'].unique()]
y = [artists.count(i) for i in x]
```

```python
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd',
          '#8c564b', '#e377c2', '#7f7f7f']
```

```python
plt.figure(figsize=(20, 20))
plt.title('Artists and their mentions in top')
plt.ylabel('mentions')
plt.bar(x[:8], y[:8], color=colors)
```

```python
wordcloud = WordCloud().generate(', '.join(x))

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud = WordCloud().generate(', '.join(data['Title']))

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
