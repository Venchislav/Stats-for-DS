```python
from sklearn import datasets
import pandas as pd
import warnings
```

```python
warnings.simplefilter('ignore')
```

```python
dataset = datasets.load_iris()
dataset
```

```python
df = pd.DataFrame(dataset['data'], columns = dataset['feature_names'])
df.head()
```

```python
df['important'] = 'No'
```

```python
df.head()
```

```python
# here we make comparisons horizontally (we sum up all columns in row that exceed mean)
comp = (df > df.mean(axis=0)).sum(axis=1)
df.loc[comp >= 3, "important"] = "Yes"
```

```python
yes = len(df[df['important'] == 'Yes'])
no = len(df[df['important'] == 'No'])
print(yes, no)
```
