```python
import pandas as pd
import scipy.stats as stats
from random import randint
import warnings
```

```python
warnings.simplefilter('ignore')
```

```python
df = pd.DataFrame({'V1':[randint(1, 10) for i in range(100)],
                   'V2':[randint(1, 10) for i in range(100)],
                   'V3':[randint(1, 10) for i in range(100)]})
df.head()
```

```python
comps = (df > df.mean(axis=0)).sum(axis=1)
df['important'] = 'No'
df.loc[comps >= (len(df.columns)-1) // 2, "important"] = "Yes"
```

```python
yes = len(df[df['important'] == 'Yes'])
no = len(df[df['important'] == 'No'])
```

```python
print(yes, no)
```
