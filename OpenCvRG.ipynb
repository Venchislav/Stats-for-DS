```python
import cv2
import matplotlib.pyplot as plt

imagePath = ['/content/ryan-gosling-blade-runner-2049-depressing-movie-scenes-hd-wallpaper-preview.jpg',
             '/content/RZMZ271_Райан_Гослинг__Ryan_Gosling__вид_5__525х1800мм_.jpg',
             '/content/59c54411-4c87-40ab-9fd6-65ef1cea-pic4_zoom-1500x1500-10701.png']
```

```python
face_classifier = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
  )
```

```python
rows = 1
columns = 3
fig = plt.figure(figsize=(6, 6))

for i in range(len(imagePath)):
  img = cv2.imread(imagePath[i])
  gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  face = face_classifier.detectMultiScale(
    gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(40, 40)
  )
  for (x, y, w, h) in face:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 4)
  img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

  fig.add_subplot(rows, columns, i+1)

  plt.imshow(img_rgb)
  plt.axis('off')
  plt.title(f"Ryan Gosling {i+1}")
```
